```python
3.6
import math 

M=6
m=6 #округление до 6 знаков после запятой
n=100 #порядок
t1=0.0001
t2=1
t3=10000

def gauss (A,b): #функция для метода Гаусса
    M=copy.copy(A) 
    s=copy.copy(b)
    x=np.zeros(n)
    
    #прямой ход
    for j in range(n):
        for i in range(j+1,n):
            mu=M[i,j]/M[j,j] #коэффициент
            M[i]-=M[j]*mu 
            s[i]-=s[j]*mu
            
    #обратный ход        
    x[n-1]=s[n-1]/M[n-1,n-1]
    
    for j in range(n-2,-1,-1):
        l=0
        for i in range(n-1-j):
            l+=(x[n-1-i]*M[j,n-1-i])
        x[j]=(s[j]-l)/M[j,j]
    return(x)

#вычисляем q по заданной формуле
q=0.993+((-1)**M)*(10**(-4))*M

#Вычислим три матрицы А для разных значений t
A_1=np.zeros([n,n])
A_2=np.zeros([n,n])
A_3=np.zeros([n,n])
for i in range(100):
    for j in range(100):
        if i!=j:
            A_1[i,j]=q**j
            A_2[i,j]=q**j
            A_3[i,j]=q**j
        else:
            A_1[i,j]=q**j+t1
            A_2[i,j]=q**j+t2
            A_3[i,j]=q**j+t3
#вычисляем вектор b по заданной формуле
b=np.zeros(n)
for j in range(n):
    b[j]=q**(n+1-j)
    
#решаем системы методом Гаусса
x_1=gauss(A_1,b)
x_2=gauss(A_2,b)
x_3=gauss(A_3,b)

#функция, округляющая число до заданной точности m 
def roundm(x):
        d = 10.0**m
        return math.floor(x*d)/d
    
#вычислим аналогично три матрицы А с округлением
A1_1=np.zeros([n,n])
A1_2=np.zeros([n,n])
A1_3=np.zeros([n,n])
for i in range(100):
    for j in range(100):
        if i!=j:
            A1_1[i,j]=roundm(q**j)
            A1_2[i,j]=roundm(q**j)
            A1_3[i,j]=roundm(q**j)
        else:
            A1_1[i,j]=roundm(q**j+t1)
            A1_2[i,j]=roundm(q**j+t2)
            A1_3[i,j]=roundm(q**j+t3)
#вектор b с округлением
b1=np.zeros(n)
for j in range(n):
    b1[j]=roundm(q**(n+1-j))
#решим полученные системы методом Гаусса    
x1_1=gauss(A1_1,b1)
x1_2=gauss(A1_2,b1)
x1_3=gauss(A1_3,b1)

#посчитаем погрешности как норму разности между полученными решениями с округлением и без округления
e1=np.linalg.norm(x_1-x1_1)
print(e1)
e2=np.linalg.norm(x_2-x1_2)
print(e2)
e3=np.linalg.norm(x_3-x1_3)
print(e3)
#чем больше величина t, тем меньше значение погрешности
```

```python
3.9
def chol(a):
  
  
    a = copy.deepcopy(a)
    
    m=a.shape[0]
    

    a[0, 0] = np.sqrt(a[0, 0])
    for i in range(1, m):
        a[i, 0] = a[i, 0] / a[0, 0]
     
    for k in range(1, m-1):
        a[k, k] = (a[k, k] - sum([a[k, q]**2 for q in range(0, k)]))**0.5
        for i in range(k+1, m):
            a[i, k] = (a[i, k] - sum([a[i, q]*a[k, q] for q in range(0, k)]))/a[k, k]
          
    a[m-1, m-1] = (a[m-1, m-1] - sum([a[m-1, q]**2 for q in range(0, m-1)]))**0.5
    
    for i in range(m):
        for j in range(i+1, m):
            a[i, j] = 0

    return a

n=20
m=8
def Aij(i, j, m=m, n=n):
    if i == j:
        return n + m**2 + j/m + i/n
    else:
        return (i+j)/(m+n)
def bi(i):
    return 200 + 50*i

A = np.array([[Aij(i, j) for i in range(1, n+1)] for j in range(1, n+1)]).reshape(n, n)
b = np.array([bi(i) for i in range(1, n+1)])
L = chol(A)
LT = L.T
y = solve(L, b)
x = solve(LT, y)

print(x, cond(A))

norm_A_rev = norm(x, ord=np.inf)/norm(b, ord=np.inf)
norm_A = norm(A, ord=np.inf)

print(norm_A*norm_A_rev)
```
